```python
# 3
import numpy as np

# Define input and output
x = 5
y = 10

# Define initial values for m and b
rand_vals=np.random.randn(3)
m,b=rand_vals[0],rand_vals[1]
print(m,b)

# Define number of iterations
num_iter = 100

# Define different learning rates
learning_rate = [0.01, 0.05, 0.1]

# Define activation functions
activation_functions = ['linear', 'sigmoid', 'relu']

# define dataframe
df = pd.DataFrame(columns=['x', 'y','y_pred','error','Learingrate','Activation'])

# Training loop
for i in range(num_iter):
    for j in range(len(learning_rate)):
        for k in range(len(activation_functions)):
            y_pred = m*x + b
            if activation_functions[k] == 'sigmoid':
                y_pred = 1 / (1 + np.exp(-y_pred))
            elif activation_functions[k] == 'relu':
                y_pred = np.maximum(0, y_pred)
            error = y - y_pred
            m = m - error * x * learning_rate[j]
            b = b - error * learning_rate[j]
            df = df.append({'x':x,'y' :y,'y_pred':y_pred,'error':error,'Learingrate' :learning_rate[j],'Activation':activation_functions[k]},ignore_index=True)
            if i % 100 == 0:
                print("Iteration: {}, Error: {}, Learning Rate: {}, Activation Function: {}".format(i, error, learning_rate[j], activation_functions[k]))

print("Final Error: {}".format(error))
```

Activation functions are mathematical functions that are applied to the output of a neuron in a neural network. They introduce non-linearity into the network, allowing the model to learn complex relationships in the data.

Activation functions are required in neural networks because without them, the network would only be able to learn linear relationships between the inputs and outputs. Activation functions allow the network to learn a wide range of functions and can help the network to converge faster.

Some popular activation functions include:

Sigmoid: It maps the input to a value between 0 and 1 and is used in the output layer of a binary classification problem.
Tanh(hyperbolic tangent): It maps the input to a value between -1 and 1 and is similar to sigmoid but it is zero-centered.
ReLU (Rectified Linear Unit): It maps negative values to 0 and positive values to the same value. It is faster to compute and can help to prevent the vanishing gradient problem.
Softmax: It maps the input values to a probability distribution and is used in the output layer of a multi-class classification problem.
Here are the Python functions for the above activation functions:



```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
```
